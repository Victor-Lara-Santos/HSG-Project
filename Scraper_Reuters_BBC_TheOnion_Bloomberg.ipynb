```python
import requests
from bs4 import BeautifulSoup

#url to be scrapped
url = 'https://www.reuters.com/news/archive/marketsNews'
result = requests.get(url)
content = result.content
soup = BeautifulSoup(content, 'html.parser')

#finding the headlines
soup.find_all('h3', {'class':'story-title'})
titles = []
for title in soup.find_all('h3', {'class':'story-title'}):
    titles.append(title.string.strip())

#subtracting number of titles so all future arrays have the same number of instances
titles = titles[:-3]

#getting the summary and preventing duplicates
first_lines = []
try:
    for first_line in soup.find_all('p'):
        first_lines.append(first_line.string.strip())
except:
    pass

#subtracting number of summaries so all future arrays have the same number of instances
first_lines = first_lines[:-2]

#getting urlsand preventing duplicates
url_links = []
import re

for tags in soup.find_all('a'):
    if re.search('article', tags['href']):
        url_links.append(tags['href'])
#subtracting number of urls so all future arrays have the same number of instances
url_links = url_links[:-6]
final_urls = []
for url in url_links:
    if url not in final_urls:
        final_urls.append(url)
import pandas as pd

reuters_word_headlines = pd.DataFrame({'title':titles, 'headline content':first_lines, 'link url':final_urls})
reuters_word_headlines=\
reuters_word_headlines.loc[:,('title', 'headline content', 'link url')]
reuters_word_headlines['link url'] =\
'www.reuters.com' + reuters_word_headlines['link url']
print(f'\n--------- Reuters ---------')
print(reuters_word_headlines)

#url to be scrapped
url = 'https://www.theonion.com/'
result = requests.get(url)
content = result.content
soup = BeautifulSoup(content, 'html.parser')

#finding the headlines
soup.find_all('h4', {'class':'sc-1qoge05-0 gWMyPL'})
titles = []
for title in soup.find_all('h4', {'class':'sc-1qoge05-0 gWMyPL'}):
    titles.append(title.string.strip())

The_Onion_headlines = pd.DataFrame({'title':titles})
The_Onion_headlines=\
The_Onion_headlines.loc[:,('title')]
print(f'\n--------- The_Onion ---------')
print(The_Onion_headlines)

headers = {'User-agent': 'Mozilla/5.0'}

request = requests.get('https://www.bbc.com/news', headers=headers)
html = request.content

soup = BeautifulSoup(html, 'html.parser')

headers = {'User-agent': 'Mozilla/5.0'}

request = requests.get('https://www.bbc.com/news', headers=headers)
html = request.content

soup = BeautifulSoup(html, 'html.parser')

def bbc_news_scraper(keyword):
    news_list2 = []

    for h in soup.findAll('h3', class_='gs-c-promo-heading__title'):
        news_title2 = h.contents[0].lower()

        if news_title2 not in news_list2:
            if 'bbc' not in news_title2:
                news_list2.append(news_title2)
# Divides BBC Headings from Reuters headings
    print(f'\n--------- BBC ---------')
    print(*news_list2, sep = '\n')
    
#in order to make the same for other websites, it is only necessary to change the html class and the heading to the specific format of the target site

#bbc_headline_scraper from indently in github was used as an inspiration and tutorial for BBC and Bloomberg scraper
#Seungjun-Data-Science in github used as inspiration for Reuters and The Onion scraper

print(bbc_news_scraper('1'))

# creates header and emulates browser
headers = {'User-agent': 'Mozilla/5.0'}

# make the request to the website in order to grab HTML
request = requests.get('https://www.bloomberg.com/markets', headers=headers)
html = request.content

soup = BeautifulSoup(html, 'html.parser')

def bloomberg_news_scraper(keyword):
    news_list = []

    # Finds all header on the website
    for h in soup.findAll('a', class_='story-package-module__story__headline-link'):
        news_title = h.contents[0].lower()

        #avoids duplicates and ads from being added to the list
        if news_title not in news_list:
            if 'bloomberg' not in news_title:
                news_list.append(news_title)
    print(f'\n--------- Bloomberg ---------')
    print(*news_list, sep = '\n')

print(bloomberg_news_scraper('1'))
```
